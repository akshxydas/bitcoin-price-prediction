```python
import numpy as np
import pandas as pd
df = pd.read_csv("bitcoin.csv")
df.head()
```

```python
df.drop(["Date"],1, inplace=True)
df.head()
```

```python
#Create new colum Prediction
df["Prediction"] = df["Price"].shift(-30)
df.head
```

```python
x = np.array(df["Price"][:-30]).reshape(-1, 1)
print(x)
```

```python
y=np.array(df["Prediction"][:-30])
y
```

```python
# Split the data into 80% training and 20% testing
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size = 0.2,random_state=99)
print(xtrain.shape, xtest.shape, ytrain.shape, ytest.shape)

predictionDays_array = np.array(df['Price'])[-30:]
print(predictionDays_array)
```

```python
from sklearn.svm import SVR
# Create and Train the Support Vector Machine (Regression) using radial basis function
svr_rbf = SVR(kernel='rbf', C=1e3, gamma=0.00001)
svr_rbf.fit(xtrain, ytrain)
```

```python
svr_rbf_confidence = svr_rbf.score(xtest,ytest)
print('SVR_RBF accuracy :',svr_rbf_confidence)
```

```python
#predict
predictions = svr_rbf.predict(xtest)
predictions
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
print("Mean Squared Error:", mean_squared_error(ytest, predictions))
print("Mean Absolute Error:", mean_absolute_error(ytest, predictions))
print("R2 Score:", r2_score(ytest, predictions))
```
